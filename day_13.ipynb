```python
def read_input(text):
    places = [i.split('\n') for i in text.split('\n\n')]
    return places
    
read_input(test_inp)[0]
```

```python
def transpose_place(place):
    return [''.join(i) for i in list(zip(*place))]

transpose_place(read_input(test_inp)[0])
```

```python
def find_reflection(place):
    for i in range(1, len(place)):
        previous = place[:i][::-1]
        following = place[i:]
        min_length = min(len(previous), len(following))
        if previous[:min_length] == following[:min_length]:
            return i
    return None
    
find_reflection(transpose_place(read_input(test_inp)[0]))
```

```python
def run(text):
    inp = read_input(text)
    s = 0
    for place in inp:
        reflextion_idx = find_reflection(place)
        if reflextion_idx is not None:
            s += 100 * reflextion_idx
        else:
            reflextion_idx = find_reflection(transpose_place(place))
            s += reflextion_idx
    return s
    
run(test_inp)
```

```python
run(inp)
```

```python
def smudge(place, coordinates):
    x, y = coordinates
    replace_value = '#' if place[x][y] == '.' else '.'
    new_place = place.copy()
    new_place[x] = new_place[x][:y] + replace_value + new_place[x][y + 1:]
    return new_place

smudge(read_input(test_inp)[0], (0, 0))
```

```python
def run2(text):
    inp = read_input(text)
    s = 0
    for i in inp:
        found = False
        old_reflextion_idx = find_reflection(i)
        if old_reflextion_idx is None:
            old_reflextion_idx = find_reflection(transpose_place(i))
        for x in range(len(i)):
            for y in range(len(i[x])):
                place = smudge(i, (x, y))
                reflextion_idx = find_reflection(place)
                reflextion_idx_trans = find_reflection(transpose_place(place))
                if reflextion_idx is not None and reflextion_idx != old_reflextion_idx:
                    found = True
                    s += 100 * reflextion_idx
                    break
                elif reflextion_idx_trans is not None and reflextion_idx_trans != old_reflextion_idx:
                    found = True
                    s += reflextion_idx_trans
                    break
            if found:
                break
    return s
    
run2(test_inp)
```

```python
run2(inp)
```
