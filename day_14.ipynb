```python
def read_input(text):
    return text.split('\n')

read_input(test_inp)
```

```python
def move_stone(dish, coord):
    x, y = coord
    if dish[x][y] == 'O' and x >= 1:
        i = x - 1
        while i >= 0 and dish[i][y] not in ['#', 'O']:
            i -= 1
        i += 1
        if i < x:
            dish[x] = dish[x][:y] + '.' + dish[x][1 + y:]
            dish[i] = dish[i][:y] + 'O' + dish[i][1 + y:]
    return dish

move_stone(read_input(test_inp), (1, 2))
```

```python
def tilt_north(dish):
    for i in range(len(dish)):
        for j in range(len(dish[i])):
            dish = move_stone(dish, (i, j))
    return(dish)

tilt_north(read_input(test_inp))
```

```python
def sum_load(dish):
    load = 0
    mult = len(dish)
    for i in range(len(dish)):
        load += mult * sum([j == 'O' for j in dish[i]])
        mult -= 1
    return load
```

```python
sum_load(tilt_north(read_input(test_inp)))
```

```python
def rotate90(dish):
    return [''.join(i) for i in list(zip(*dish))[::-1]]

def rotate180(dish):
    return rotate90(rotate90(dish))

def rotate270(dish):
    return rotate90(rotate180(dish))

def tilt_west(dish):
    dish = rotate270(dish)
    dish = tilt_north(dish)
    dish = rotate90(dish)
    return dish

def tilt_south(dish):
    dish = rotate180(dish)
    dish = tilt_north(dish)
    dish = rotate180(dish)
    return dish

def tilt_east(dish):
    dish = rotate90(dish)
    dish = tilt_north(dish)
    dish = rotate270(dish)
    return dish
```

```python
def find_pattern(vals):
    layers = [vals]
    steps = 0
    
    while sum([abs(i) for i in layers[-1]]) != 0:
        new_layer = [layers[-1][i + 1] - layers[-1][i] for i in range(len(layers[-1]) - 1)]
        layers.append(new_layer)
        steps += 1
    return layers, steps

find_pattern([10, 13, 16, 21, 30, 45])
```

```python
def run_part_2(dish, cycles):
    s = 0
    pattern = []
    for i in range(1, cycles + 1):
        dish = tilt_north(dish)
        dish = tilt_west(dish)
        dish = tilt_south(dish)
        dish = tilt_east(dish)
        pattern.append(sum_load(dish))
        print(i, sum_load(dish))
    load = sum_load(dish)
    return load
        
run_part_2(read_input(test_inp), 40)
```
