```python
def read_inp(text):
    instructions = text.split('\n')
    instructions = [re.findall(r'([LRUD]) (\d+)', i)[0] for i in instructions]
    instructions = [(i[0], int(i[1])) for i in instructions]
    return instructions

read_inp(test_inp)
```

```python
def dig_trench(instructions):
    trench = ['#']
    x = 0
    y = 0
    for i in instructions:
        height = len(trench)
        width = len(trench[0])
        d, n = i
        if d == 'L':
            diff = n - y
            for j in range(len(trench)):
                trench[j] = '.' * diff + trench[j]
            trench[x] = trench[x][: y - n] + '#' * n + trench[x][y:]
            y -= n
        elif d == 'R':
            diff = y - width + n + 1
            for j in range(len(trench)):
                trench[j] = trench[j] + '.' * diff
            trench[x] = trench[x][: y + 1] + '#' * n + trench[x][y + n + 1:]
            y += n
        elif d == 'U':
            diff = n - x
            for j in range(diff):
                trench.insert(0, '.' * width)
            for j in range(n):
                trench[x - j - 1] = trench[x - j - 1][: y] + '#' + trench[x - j - 1][y + 1:]
            x -= n
        elif d == 'D':
            diff = x - height + n + 1
            for j in range(diff):
                trench.append('.' * width)
            for j in range(n):
                trench[x + j + 1] = trench[x + j + 1][: y] + '#' + trench[x + j + 1][y + 1:]
            x += n
    return trench

dig_trench(read_inp(test_inp))
```

```python
def fill_between(text):
    c = 0
    starts = []
    ends = []
    for i in re.finditer(r'#+', text):
        if c % 2 == 0:
            starts.append(i.end())
        elif c % 2 == 1:
            ends.append(i.start() + 1)
        c += 1
    for i in range(len(starts)):
        if len(ends) >= i + 1:
            start = starts[i]
            end = ends[i]
            text = text[:start] + '#' * (end - start) + text[end:]
    return text

fill_between('##...###.....#.#')
```

```python
def dig_interior(trench):
    new_trench = trench.copy()
    for i in range(len(new_trench)):
        new_trench[i] = fill_between(new_trench[i])
    return new_trench

dig_interior(dig_trench(read_inp(test_inp)))
```

```python
def get_area(trench):
    s = 0
    for i in trench:
        s += sum([j == '#' for j in i])
    return s
    
get_area(dig_interior(dig_trench(read_inp(test_inp))))
```
