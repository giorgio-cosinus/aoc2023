```python
def read_input(text):
    garden = text.split('\n')
#     garden = [list(i) for i in garden]
    return garden
    
read_input(test_inp)
```

```python
def get_available_neighbors(positions, coordinates):
    neighbors = {}
    for i in range(4):
        left_coord = (coordinates[0] - 1, coordinates[1])
        right_coord = (coordinates[0] + 1, coordinates[1])
        up_coord = (coordinates[0], coordinates[1] - 1)
        down_coord = (coordinates[0], coordinates[1] + 1)
        if positions.get(left_coord) == '.':
            neighbors[left_coord] = positions.get(left_coord)
        if positions.get(right_coord) == '.':
            neighbors[right_coord] = positions.get(right_coord)
        if positions.get(up_coord) == '.':
            neighbors[up_coord] = positions.get(up_coord)
        if positions.get(down_coord) == '.':
            neighbors[down_coord] = positions.get(down_coord)
    return neighbors
```

```python
def take_step(garden):
    positions = {}
    new_garden = garden.copy()
    for i, v in enumerate(garden):
        for j, w in enumerate(v):
            positions[(i, j)] = w
    for k, v in positions.items():
        if v in ['O', 'S']:
            new_garden[k[0]] = new_garden[k[0]][:k[1]] + '.' + new_garden[k[0]][k[1] + 1:]
            available_neighbors = get_available_neighbors(positions, k)
            for l, w in available_neighbors.items():
                if w == '.':
                    new_garden[l[0]] = new_garden[l[0]][:l[1]] + 'O' + new_garden[l[0]][l[1] + 1:]
    return new_garden

take_step(read_input(test_inp))
```

```python
def take_steps(garden, steps):
    new_garden = garden.copy()
    for i in range(steps):
        new_garden = take_step(new_garden)
    return new_garden

take_steps(read_input(test_inp), 6)  
```

```python
def count_positions(garden):
    s = 0
    for i in garden:
        s += sum([j == 'O' for j in list(i)])
    return s

count_positions(take_steps(read_input(test_inp), 6))
```
